```python
# -*- coding: utf-8 -*-

import re
import requests
from bs4 import BeautifulSoup as bs

head = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.73 Safari/537.36'}

def mv_name(soup, name=None):#电影名字
    span = soup.select('.title')
    for title in span:
        if title.string[1] != '/': #去掉电影的英文名
            if name is None:
                name = []
            name.append(title.string)
            
    return name

def mv_eval_num(soup): #电影评价人数
    mv_eval = soup.select('.star')
    eval_num = re.findall(r'<span>(.*?)</span>', str(mv_eval), re.S)
    return eval_num

def mv_score(soup): # 电影评分
    p_score = soup.select('.star')
    movie_score = re.findall(r'<span class="rating_num" property="v:average">(.*?)</span>', str(p_score), re.S)
    return movie_score

def mv_quote(soup, l=None): #一句概括电影的话
        p_quote = soup.select('.quote') 
        movie_quote = re.findall(r'<span class="inq">(.*?)</span>', str(p_quote), re.S)
        for quote_info in movie_quote:
            if l is None:
                l = []
            l.append(quote_info)
            
        return l

def mv_dr_ar(soup, Info=None): #导演和演员信息
    dr_aror_info = soup.select('p')
    dr_ar_info = re.findall(r'<p class="">(.*?)<br/>', str(dr_aror_info), re.S)
    for info in dr_ar_info:
        if Info is None:
            Info = []
        Info.append(info.strip()) #strip()是为了去掉多余的空格
        
    return Info

if __name__ == '__main__':
    
    for i in range(10): #获得全部电影信息
        url = 'https://movie.douban.com/top250?start=%d&filter='%(i*25)
        res = requests.get(url, headers=head)
        soup = bs(res.text)
        name = mv_name(soup, name=None)
        score = mv_score(soup)
        quote = mv_quote(soup)
        num = mv_eval_num(soup)
        dr_ar = mv_dr_ar(soup)
        
        #'a' 自动创建以douban004.csv文件，以追加的方式写入，若文件已存在，则不写入
        with open('/home/mouse/anaconda3/dbmvtop250.csv', 'a') as f: #保存路径改成自己的
            for j in range(len(name)):
                movies_infos = '%s; %s ;%s; %s; %s' % (name[j], num[j], score[j], dr_ar[j], quote[j]) 
                #上面加';'是为了写入csv文件中让每个信息单格显示
                f.writelines(movies_infos + '\n')
                print('Downloading.............')
                
    print('Crawl succeed')
```
